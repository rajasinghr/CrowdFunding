```python
import PyPDF2
import os 

pdfSourceFolder = r'../../Crowdfunding Papers-20190208T234520Z-001/Crowdfunding Papers/' 
txtDestFolder = r'../../Crowdfunding Papers-20190208T234520Z-001/Crowdfunding Papers - References/' 
pdfFiles = []
if not os.path.exists(txtDestFolder):
    os.makedirs(txtDestFolder)
for filename in os.listdir(pdfSourceFolder):
    if filename.endswith('.pdf'):
        pdfFiles.append(filename)
```

```python
for pdfFile in pdfFiles:
    read_pdf = PyPDF2.PdfFileReader(open(pdfSourceFolder+pdfFile,'rb'))
    number_of_pages = read_pdf.getNumPages()
    f = open(txtDestFolder+pdfFile.replace('.pdf','.txt'),'w+',encoding='utf-8')
    page_content = ''
    for page in range(0,number_of_pages):
        page = read_pdf.getPage(page)
        page_content += page.extractText()
    refTextSplit = page_content.lower().replace('\n','').split('references')
    f.write(refTextSplit[len(refTextSplit)-1])
    f.write("\r\n")
    f.close()
```
